```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Output File (CSV)
master_data_file = 'Data/chicago_data.csv'

#import data set ids for the https://data.cityofchicago.org/resource/
data_input = pd.read_csv('Data/data_ids.csv')
data_ids = data_input['1'].tolist()

#Data request url
url = f"https://data.cityofchicago.org/resource/"
par = ".json"

#import shell data value dataframe
city_data = pd.read_csv('Data/df_start.csv')

#API calls to import each data set and merge it based on date values
for item in data_ids:
    print(url + item + par)
    response = requests.get(url + item + par).json()
    new_data = pd.DataFrame(response)
    new_data.columns.values[0] = 'date'
    city_data = pd.merge(city_data, new_data, how='outer', on='date')
```

```python
#Save the merged dataset as a .csv and display the dataframe
city_data.to_csv(master_data_file, index=False)
city_data
```
