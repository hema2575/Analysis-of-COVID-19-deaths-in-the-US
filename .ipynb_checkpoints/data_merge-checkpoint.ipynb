```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Output File (CSV)
master_data_file = 'Data/chicago_data.csv'

#import data set ids for the https://data.cityofchicago.org/resource/
data_input = pd.read_csv('Data/data_ids.csv')
data_ids = data_input['1'].tolist()
print(data_ids)
#Data request url
url = f"https://data.cityofchicago.org/resource/"
par = ".json"

#import shell data value dataframe
city_data = pd.read_csv('Data/df_start.csv')

#API calls to import each data set and merge it based on date values
for item in data_ids:
#    print(url + item + par)
    response = requests.get(url + item + par).json()
    new_data = pd.DataFrame(response)
    new_data.columns.values[0] = 'date'
    city_data = pd.merge(city_data, new_data, how='outer', on='date')
```

```python
#Save the merged dataset as a .csv and display the dataframe
city_data.to_csv(master_data_file, index=False)
city_data
```

```python
city_data = city_data.dropna(subset=['date','hospitalizations_total'])
city_data.to_csv(master_data_file, index=False)
city_data
```

```python
import scipy.stats as stats
racedf = pd.DataFrame(city_data[['cases_latinx','cases_asian_non_latinx','cases_black_non_latinx','cases_white_non_latinx','cases_other_non_latinx','cases_unknown_race_eth']])
#print(racedf)
#print(type(racedf.iloc[1]['cases_latinx'])) 
racedf = racedf.astype('int64')
race_sum = pd.DataFrame(racedf.sum()).rename(columns={0:'observed'})
race_sum.drop(['cases_unknown_race_eth'], inplace = True)
observed_total = race_sum['observed'].sum()
race_sum['observed_%'] = race_sum['observed']/observed_total *100
race_sum['expected'] = [776290,184768,768524,901941,62432]
chicago_total = race_sum['expected'].sum()
race_sum['expected_%'] = race_sum['expected']/chicago_total *100
race_sum
```

```python
critical_value = stats.chi2.ppf(q = 0.95, df = 4)
critical_value
```

```python
stats.chisquare(race_sum['observed_%'], race_sum['expected_%'])
```
