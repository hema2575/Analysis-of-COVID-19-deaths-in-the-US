```python
#### Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os

# Import US Pre Pandemic Bed Data File
beddata_path = os.path.join('USBedsData-PrePandemic','Definitive_HealthcareUSA_Hospital_Beds.csv')
df = pd.read_csv(beddata_path)
beddata_path
df
```

```python
df.groupby('HQ_STATE').sum()
```

```python
#Creata bar chart to see the Number of ICU Beds & Bed Utilization in the United States
state_df = df.groupby('HQ_STATE').sum()[['NUM_ICU_BEDS','BED_UTILIZATION','Potential_Increase_In_Bed_Capac']]
ax = state_df.plot.bar(stacked=True)
plt.title("The Utilization of Hosiptal Beds in the United States")
plt.xlabel("States")
plt.ylabel("Number of Beds")
plt.legend(loc="lower left", bbox_to_anchor=(1.1, 0.2), fontsize='small', title='Hosiptal Beds')
#Increase Figure Size
plt.figure(figsize=(100, 80))
#Increase x-axis so we can be able to see the states 
```
